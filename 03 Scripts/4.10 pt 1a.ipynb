## creating our Region column

```python
#I'm thinking a for loop would be best here because we only want it to cycle through one column
#first, we want to know what values we're working with

df['state'].value_counts()
```

```python
#making the loop
region = []

for value in df['state']:
    if value == 'Maine' or value == 'New Hampshire' or value == 'Vermont' or value == 'Massachusetts' or value == 'Connecticut' or value == 'New York' or value == 'Pennsylvania' or value == 'New Jersey':
        region.append('Northeast')
    elif value == 'Wisconsin' or value == 'Michigan' or value == 'Indiana' or value == 'Illinois' or value == 'Ohio' or value == 'North Dakota' or value == 'South Dakota' or value == 'Kansas' or value == 'Nebraska' or value == 'Minnesota' or value == 'Iowa' or value == 'Missouri':
        region.append('Midwest')
    elif value == 'Delaware' or value == 'Maryland' or value == 'District of Columbia' or value == 'Virginia' or value == 'West Virginia' or value == 'North Carolina' or value == 'South Carolina' or value == 'Georgia' or value == 'Florida' or value == 'Kentucky' or value == 'Tennessee' or value == 'Mississippi' or value == 'Alabama' or value == 'Oklahoma' or value == 'Texas' or value == 'Arkansas' or value == 'Louisiana':
        region.append('South')
    else:
        region.append('West')
```

```python
#adding our new column
df['region'] = region
```

## making customer profiles

I think it's going to take some further exploration of the data to really get a handle on different profiles. I'm going to start with age, # of dependents, and income only, then we'll see how the rest break down. I tried to think of groups who were most likely to be using Instacart, as it's an obvious choice for a wealthy family juggling 6 kids but might not appeal as much to an older, blue-collar or traditional style couple.

```python
#we're going to make our own function here, since we need to go through the whole data but also an 'other' column for customers who don't fit these profiles
def profile(row):

  if row['age'] >= 18 & row['age'] < 55 & row['income'] <= 50000 & row['n_dependents'] >= 1:
    return 'Low income family'
#if both parents have to work to support their family, they likely want the convenience of a delivery app
  elif row['age'] >= 18 & row['age'] < 55 & row['income'] >= 100000 & row['n_dependents'] >= 1:
    return 'Wealthy family'
#a wealthy family can afford to use this kind of service, so why wouldn't they?
  elif row['age'] >= 27 & row['age'] < 42 & row['annual_income'] < 50000 & row['n_dependents'] < 1:
    return 'Struggling millenial'
#millenials as a generation seem to love this kind of thing
  elif row['age'] >= 30 & row['age'] < 60 & row['annual_income'] > 75000 & row['n_dependents'] < 1:
    return 'Successful GenXer/empty nester'
#If they're busy with their careers and don't have to think too hard about meal planning for kids, etc. this is very convenient
  elif row['age'] >= 60 & row['n_dependents'] < 1:
    return 'Senior citizen'
#it's way more of an endeavor for an elderly person to go to the supermarket, so this is a good choice for them
  else: return 'Other'
#not everyone will fit into a distinct category like this obviously, I just tried to picture the most likely groups to use Instacart
```

(question: this function does not work if I set the n_dependents equal to 0, but does work if I set it to 'less than 1'. Why is this?)

```python
#now we make the new column that applies this function
df2['customer_profile'] = df2.apply(profile, axis=1)
```
